```python
stu1=[56,78,92,56,90,89]
stu2=[13,34,54,23,30,34]
sub=['maths','science','socia','english','telugu','hindi']
plt.plot(sub,stu1,label='student1',color='k')
plt.plot(sub,stu2,label='student2',color='r')
plt.legend()
```

```python
stu1=[56,78,92,56,90,89]
stu2=[13,34,54,23,30,34]
sub=['maths','science','socia','english','telugu','hindi']
plt.subplot(2,1,1)
plt.plot(sub,stu1,label='student1',color='k')
plt.legend()
plt.subplot(2,1,2)
plt.plot(sub,stu2,label='student2',color='r')
plt.legend()
```

```python
year=np.array(['19-20','20-21','21-22','22-23'])
comA=np.array([230,560,780,127,128])
comB=np.array([200,160,270,127,400])
print()
proA=np.diff(comA)
proB=np.diff(comB)
print(proA)
print(proB)
plt.subplot( 1,2,1)
plt.bar(year,proA,label='A')
plt.legend(loc="best")
plt.subplot(1,2,2)
plt.bar(year,proB,label='B')
plt.legend(loc="best")
```
